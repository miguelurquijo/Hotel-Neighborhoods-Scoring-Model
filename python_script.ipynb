# Peer-graded Assignment: Segmenting and Clustering Neighborhoods in Toronto

## 1 Adjust the DataFrame

### 1.1 Scrape the Wikipedia page with Pandas

```python
Postal_Codes = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
Postal_Codesdf = Postal_Codes[0]
Postal_Codesdf.head(15)
```

### 1.2 Lets drop the Borough with value = 'Not assigned'

```python
i = Postal_Codesdf[((Postal_Codesdf.Borough == 'Not assigned'))].index
Postal_Codesdf = Postal_Codesdf.drop(i)
Postal_Codesdf.head(15)
```

### 1.3 Group by Borough

```python
pd.set_option('max_colwidth', 100)
Postal_Codesdf = Postal_Codesdf.groupby(['Postal code'], as_index=False).first() #Group by postal code 
Postal_Codesdf['Neighborhood'] = Postal_Codesdf['Neighborhood'].str.replace('/',',') #replace '/' by ','
Postal_Codesdf.head(15)
```

```python
Postal_Codesdf.shape
```

## 2 Get the coordinates of each Postal Code

### 2.1 Get the Latittude and Longitude in a DataFrame

```python
coordinates = pd.read_csv('http://cocl.us/Geospatial_data')
coordinates.head(15)
```

### 2.2 Get the Latittude and Longitude in the Postal Codes Data Frame

```python
Postal_Codesdf = Postal_Codesdf.set_index('Postal code').join(coordinates.set_index('Postal Code'))
Postal_Codesdf
```

### 3.2 Explore and cluster the neighborhoods in Toronto

```python
#Filter the DataFrame with Toronto locations
Toronto = Postal_Codesdf[Postal_Codesdf['Borough'].str.contains('Toronto')]
Toronto.head(15)
```

```python
toronto_latitude = 43.6532
toronto_longitude = -79.3832
map_toronto = folium.Map(location=[toronto_latitude, toronto_longitude], zoom_start=12)
```

```python
for lat, lng, borough, neighborhood in zip(Toronto['Latitude'], Toronto['Longitude'], Toronto['Borough'], Toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
